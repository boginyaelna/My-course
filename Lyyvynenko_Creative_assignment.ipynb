```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('yulu_rental.csv')
```

Питяння перше: Дослідження впливу температури, вологості та швидкості вітра на кількість орендованих велосипедів. 
Побудуємо три діаграми розсіювання для кожної з досліджуваніх категорій і загальної кількості оренди.

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
sns.scatterplot(data=df, x='temp', y='count', hue='temp', palette='coolwarm')
plt.title('Температура vs Кількість оренд')
plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='humidity', y='count', hue='humidity', palette='Blues')
plt.title('Вологість vs Кількість оренд')
plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x='windspeed', y='count', hue='windspeed', palette='YlGnBu')
plt.title('Швидкість вітра vs Кількість оренд')
plt.tight_layout()
plt.show();
```

1. Температура vs Кількість оренд (Перша діаграма)
	- Позитивна залежність: Зі збільшенням температури кількість оренд зростає. Низькі температури (від 0 до 15°C) мають менше оренд, тоді як у діапазоні 20-35°C оренди значно збільшуються.
	- Оптимальні температури: Найбільше оренд зафіксовано в середньому діапазоні температури (~20-35°C), що можна пояснити комфортними умовами для поїздок на велосипеді.
	- Високі температури (>35°C): Починається спад кількості оренд, ймовірно, через надмірну спеку, що не сприяє активному використанню велосипедів.
2. Вологість vs Кількість оренд (Друга діаграма)
	- Негативний вплив високої вологості: При підвищенні вологості понад 60-80% кількість оренд має тенденцію знижуватись. Це може бути пов’язано з дискомфортом під час катання у вологому середовищі.
	- Низька вологість (<40%): Спостерігається більш висока активність оренди велосипедів, оскільки умови є сухими та сприятливими для катання.
	- Широкий розкид: Вологість до 100% не завжди різко знижує кількість оренд, але загальна тенденція показує спад.
3. Швидкість вітру vs Кількість оренд (Третя діаграма)
	- Низька швидкість вітру (0-20 км/год): Велика кількість оренд відбувається при низькій швидкості вітру. Це логічно, адже сильний вітер ускладнює катання.
	- Середня швидкість (20-30 км/год): Оренди знижуються, але все ще присутні у значній кількості.
	- Висока швидкість вітру (>30 км/год): Кількість оренд помітно зменшується, оскільки сильний вітер створює фізичний дискомфорт та небезпеку для велосипедистів.
	Рекомендації для бізнесу:
	- Оптимізація пропозиції: Фокус на маркетингові кампанії під час сприятливих погодних умов (помірна температура і низька вологість).
	- Погодні прогнози: Використовувати прогнози для коригування наявності велосипедів і обслуговування у несприятливі дні.
	- Комфортні послуги: У регіонах з високою вологістю чи вітряністю можна пропонувати додаткові рішення, такі як велосипедні парковки чи спорядження для захисту від погодних умов.

Друге питання: чи є відмінності в кількості оренд у різні погодні умови (weather) залежно від сезону (season)?
Побудуемо boxplot, де x='season' — категорії погоди на осі X, y='count' — кількість оренд на осі Y, hue='weather' — розрізняє пори року за кольорами.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df,x='season', y='count', hue='weather', palette='Set2')
plt.title('Розподіл кількості оренд за сезонами залежно від погодних умов')
plt.xlabel('Сезон')
plt.ylabel('Кількість оренд')
plt.legend(title='Погодні умови', loc='upper center', bbox_to_anchor=(1.15, 0.5))
plt.show();
```

1. Загальна тенденція:
	- Позитивні погодні умови (умова 1) забезпечують найвищу кількість оренд у всіх сезонах, що є очікуваним, оскільки користувачі віддають перевагу велосипедам у ясну або хмарну погоду.
	- У поганих погодних умовах (умови 3) кількість оренд різко зменшується — цей тренд можна чітко побачити по висоті стовпців.
    - Умови 4 (сильний дощ/сніг) мають найнижчі показники, і вони спостерігаються лише у весняний сезон (сезон 1).
2.	Вплив сезонів:
	- Літо (сезон 2): Найбільша кількість оренд, як за середніми значеннями, так і за максимальними піками. Це пояснюється сприятливою погодою та високою активністю користувачів.
	- Осінь (сезон 3): Теж демонструє високі показники оренди, хоча трохи нижчі за літо.
	- Зима (сезон 4): Має нижчі середні показники, але присутні викиди, що свідчать про певні дні підвищеного попиту.
	- Весна (сезон 1): Найменша кількість оренд серед усіх сезонів. Особливо низькі значення спостерігаються при поганих погодних умовах.
3. Викиди:
    - Верхні “вусики” та точки-викиди на діаграмі свідчать про аномально високі значення кількості оренд у певні дні. Навіть у несприятливих погодних умовах, таких як умови 3 (дощ) чи умови 4 (сильний дощ/сніг), присутні викиди. Це означає, що деякі окремі дні демонструють аномально високий попит, імовірно через локальні події або свята, коли навіть погана погода не заважає активності користувачів або нестандартну поведінку користувачів.

Третє питання друге: Дослідимо кількість зареестрованих і незареестрованих користувачів погодинно.
Змінимо тип даних колонки datetime на pandas.datetime. Потім витягнемо години та згрупуємо по них і побудуемо прості лінійни графіки.

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df['hour'] = df['datetime'].dt.hour
```

```python
df_hour = df[['hour', 'registered', 'casual']].groupby('hour').sum()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_hour.index, df_hour['registered'], label='Зареєстровані користувачі', color='#2ca02c', marker='o', linewidth=2)
plt.plot(df_hour.index, df_hour['casual'], label='Незареєстровані користувачі', color='#ff7f0e', marker='o', linewidth=2)
plt.title('Порівняння зареєстрованих та незареєстрованих користувачів по годинах', fontsize=16)
plt.xlabel('Година доби', fontsize=14)
plt.ylabel('Кількість користувачів', fontsize=14)
plt.legend()
plt.grid(True, which='both', axis='both', linewidth=0.5, alpha=0.7)
plt.xticks(range(0, 24))  
plt.tight_layout()
plt.show();
```

Аналітичні висновки:
	1.	Ранок (6:00-9:00):
Спостерігається різке зростання кількості користувачів. Найбільший приріст у 8:00. Це піковий час для обох груп користувачів, що може свідчити про робочі поїздки та використання транспорту для початку дня.
	2.	Обідній час (12:00-14:00):
Кількість зареєстрованих користувачів стабільно висока, що вказує на активність під час обідньої перерви. Незареєстровані користувачі також активно користуються послугами.
	3.	Вечірній пік (16:00-18:00):
Найвища активність спостерігається о 17:00. Зареєстровані користувачі мають максимальний показник (179356), що пояснюється завершенням робочого дня та поверненням додому. Незареєстровані користувачі також демонструють активність.
	4.	Нічний період (0:00-5:00):
Активність значно знижується в нічні години, що є логічним, оскільки більшість користувачів не користуються послугами в цей час.
	5.	Порівняння груп користувачів:
Зареєстровані користувачі значно переважають незареєстрованих протягом усього дня, особливо у пікові години. Це свідчить про велику кількість постійних користувачів сервісу.
