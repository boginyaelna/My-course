Ваша фінальна структура (запишіть її в змінну `df_registered_by_date`) даних має містити дві колонки: дату і кількість зареєстрованих користувачів на цей день.

```python
df['date'] = df.datetime.dt.date
```

```python
df_registered_by_date = df.groupby('date', as_index=False)['registered'].sum()
df_registered_by_date.head()
```

5. Тепер побудуйте лінійний графік з matplotlib для даних в `df_registered_by_date`.

  Додайте наступну частину до графіку аби зробити підписи за віссю ОХ більш читаємими

  `plt.xticks(rotation=45, ha='right')`

  Після побудови графіку проаналізуйте його та напишіть, які тенденції щодо зміни кількості зареєстрованих користувачів в часі ви бачите і чому так може бути?

```python
x = df_registered_by_date.date
y = df_registered_by_date.registered
plt.plot(x,y, 'orchid')
plt.title('Поденна регістрація', color = 'indigo')
plt.xlabel('Дата', color = 'darkviolet')
plt.ylabel('Кількість зареєстрованих користувачів', color = 'darkviolet')
plt.xticks(rotation=45, ha='right')
plt.tight_layout();
```

З графіка наглядно видно наступні тенденції:
1. Загальний ріст кількості користувачів:

	•	Компанія активно розширювала свою зону покриття, проводила маркетингові кампанії та скоріш за все збільшувала кількість станцій із велосипедами.
2. Коливання та сезонність:
	•	Чітко спостерігаються коливання в кількості зареєстрованих користувачів, що пов’язано із сезонними змінами. Наприклад, влітку попит зростав через зручні погодні умови для використання велосипедів.

6. Побудуйте графік розсіювання (scatterplot) між температурою і тим, як вона відчувається (atemp). Поставте прозорість точок 0.2. Додайте підписи до осей Х ат У. Опишіть свої спостереження з цього графіку. Які висновки можна зробити? Що дивного бачите?

```python
x = df.temp
y = df.atemp
plt.scatter(x,y,color = 'slateblue', alpha=0.2)
plt.title('Відчуття температури')
plt.xlabel('Реальний показник температури', color = 'indigo')
plt.ylabel('Як відчувється температура', color = 'indigo')
plt.tight_layout();
```

Мої спостереження:
1.	Чітка лінійна залежність: Значення мають майже ідеальну лінійну форму, що свідчить про сильний кореляційний зв’язок.
2.	Незвичайні точки: На графіку є горизонтальний кластер точок (температура приблизно 25°C), де “відчувається температура” змінюється, але значення реальної температури залишається постійним. Це може свідчити про проблеми з даними або якісь специфічні кліматичні умови.
